```python
import pandas as pd
myData = pd.read_csv('spider04_forClassifyMyMake.csv',usecols = ['YingYu','ShuXue','Class'])
myData.head()
```

```python
from sklearn.model_selection import train_test_split
trainSet_x,testSet_x,trainSet_y,testSet_y=train_test_split(myData.iloc[:,0:2],myData.iloc[:,2],test_size=0.2,random_state=22)
```

```python
trainSet_x.shape
```

```python
from sklearn.model_selection import cross_val_score
def fit_model(model,trainSet_x,testSet_x,trainSet_y,testSet_y):
    model=model.fit(trainSet_x,trainSet_y)
    predict=model.predict(testSet_x)
    print("%s score is : %0.2f" % ("model",model.score(testSet_x,testSet_y)))
    scores=cross_val_score(model,myData.iloc[:,0:2],myData.iloc[:,2],cv=5)
    print('%s avgscore is : %0.2f(+/- %0.2f)' % ('model',scores.mean(),scores.std()))
    return(model)
```

```python
from sklearn.tree import DecisionTreeClassifier
model001= DecisionTreeClassifier(criterion="entropy",max_depth=5)

model001=fit_model(model001,trainSet_x,testSet_x,trainSet_y,testSet_y)
```

```python
import numpy as np
#画分类图
import matplotlib as mpl
import matplotlib.pyplot as plt
def plot(train_x,train_y,test_x,test_y,model):
    x1_min, x1_max = train_x.iloc[:, 0].min(), train_x.iloc[:, 0].max()  # yingyu的最低分和最高分 
    x2_min, x2_max = train_x.iloc[:, 1].min(), train_x.iloc[:, 1].max()  # shuxue的最低分和最高分
    x1, x2 = np.mgrid[x1_min:x1_max:80j, x2_min:x2_max:80j]  # 生成网格采样点(根据x1与x2的范围即56~96)  200j表示精度，越大越准确。步长为复数表示点数（取200个），左闭右闭；步长为实数表示间隔，左闭右开
    grid_test = np.stack((x1.flat, x2.flat), axis=1)           # 测试点  # flat是将数组转换为一维迭代器,按照第一维将x1,x2进行拼合.stack函数用于堆叠数组，取出第二维（axis=1纵轴，若=0，第一维）进行打包；              
    grid_hat = model.predict(grid_test)                          # 预测分类值  
    grid_hat = grid_hat.reshape(x1.shape)                      # 使之与输入的x1形状相同，转换为200*200,转换前为40000*1

    cm_light = mpl.colors.ListedColormap(['lightgreen', 'pink', 'lightblue'])  #绿 红 蓝      (['#A0FFA0', '#FFA0A0', '#A0A0FF'])
    cm_dark = mpl.colors.ListedColormap(['g', 'r', 'b']) 
    cm_dark1= mpl.colors.ListedColormap(['darkgreen', 'darkred', 'darkblue']) 
    plt.pcolormesh(x1, x2, grid_hat, cmap=cm_light)            # 画分类图，绘制背景
    plt.scatter(train_x.iloc[:, 0], train_x.iloc[:, 1], c=train_y, edgecolors='black', s=20, cmap=cm_dark)  # 画样本点, c=y表示按照y的值来区分（x,y）点的颜色，edgecolors是指描绘点的边缘色彩-黑色，s指描绘点的大小，cmap指点的颜色
    plt.scatter(test_x.iloc[:, 0], test_x.iloc[:, 1], c=test_y, edgecolors='black', s=40, cmap=cm_dark1)    # 圈中测试集样本
    plt.xlabel('English', fontsize=13)  
    plt.ylabel('Math', fontsize=13)  
    plt.xlim(x1_min, x1_max)  
    plt.ylim(x2_min, x2_max)  
    plt.title('Student\'s Grade', fontsize=15)   
    plt.show()
```

```python
plot(trainSet_x,trainSet_y,testSet_x,testSet_y,model001)
```

```python

def tree_typed_figure(model):
    dot_data = tree.export_graphviz(
        model,
        out_file = None,
        feature_names = trainSet_x.columns.values,
        filled = True,  #彩色的
        impurity = True,
        rounded = True   #输出的结点四个角是圆润的，False则是直角
   )
    return(dot_data)
```

```python
import pydotplus
from sklearn import tree
from IPython.display import Image
dot_data=tree_typed_figure(model001)
graph=pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```

```python
from sklearn.ensemble import RandomForestClassifier
model002=RandomForestClassifier(random_state=42,max_depth=5)
model002=fit_model(model002,trainSet_x,testSet_x,trainSet_y,testSet_y)
plot(trainSet_x,trainSet_y,testSet_x,testSet_y,model002)
```
